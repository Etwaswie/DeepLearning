```python
import json

with open('/content/class_map.json', 'r') as f:
    class_map = json.load(f)
```

```python
groups = class_map['groups']
sorted(groups)
```

```python
group_map = {g: i for i, g in enumerate(sorted(groups))}
group_map
```

```python
new_class_map = {}
for i, k in enumerate(class_map['label_map'].keys()):
    new_class_map[i] = group_map[class_map['label_map'][k]]
```

```python
import os

root_dirs = ['/content/russian-traffic-signs-recognition-1/train/labels',
             '/content/russian-traffic-signs-recognition-1/valid/labels',
             '/content/russian-traffic-signs-recognition-1/test/labels']
for root in root_dirs:
    labels = os.listdir(root)
    for l in labels:
        label_file = root + '/' + l
        with open(label_file, 'r') as f:
            annot = f.readlines()

        new_annot = []
        for a in annot:
            a = a.rstrip().split(' ')
            lbl = new_class_map[int(a[0])]
            new_annot.append(f"{lbl} {' '.join(a[1:])}")
        with open(label_file, 'w') as f:
            for a in new_annot:
                f.write(a + '\n')
```
